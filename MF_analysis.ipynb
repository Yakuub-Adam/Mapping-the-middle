```python
# Step 1: Upload the file
from google.colab import files
uploaded = files.upload()  # This opens a file picker

# Step 2: Create a DataFrame from the uploaded file
import pandas as pd

# Use the actual filename as it appears in uploaded.keys()
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

# Step 3: Check the first few rows to make sure it loaded correctly
df.head()
```

```python
df_mf = df[df['Pos'] == 'MF']
df_mf.head()
```

```python
defensive_metrics = ['Player', 'Squad', 'Tkl', 'Int', 'Tkl+Int']
df_defensive = df_mf[defensive_metrics]
df_defensive.head()
```

```python
top_tackles = df_defensive[['Player', 'Squad', 'Tkl']].sort_values(by='Tkl', ascending=False).head(3)
print("Top 3 Tackles:\n", top_tackles)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.bar(top_tackles['Player'], top_tackles['Tkl'], color='royalblue', alpha=0.8)
plt.title("Top 3 Midfielders : Tackles", fontsize=14, weight='bold')
plt.ylabel("Total Tackles", fontsize=12)
plt.xlabel("Player", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
top_interceptions = df_defensive[['Player', 'Squad', 'Int']].sort_values(by='Int', ascending=False).head(3)
print("\nTop 3 Interceptions:\n", top_interceptions)
```

```python
plt.figure(figsize=(8,5))


plt.barh(top_interceptions['Player'], top_interceptions['Int'], color='seagreen', alpha=0.7)


plt.gca().invert_yaxis()

plt.xlabel("Interceptions")
plt.title("Top 3 Midfielders: Interceptions", fontsize=14, weight='bold')



plt.show()
```

```python
top_combined = df_defensive[['Player', 'Squad', 'Tkl+Int']].sort_values(by='Tkl+Int', ascending=False).head(3)
print("\nTop 3 Combined Tackles + Interceptions:\n", top_combined)
```

```python
import matplotlib.pyplot as plt


top_xAG = df_mf[['Player', 'Squad', 'xAG']].sort_values(by='xAG', ascending=False).head(3)


plt.figure(figsize=(8,5))
plt.barh(top_xAG['Player'] + " (" + top_xAG['Squad'] + ")", top_xAG['xAG'], color='royalblue')

plt.gca().invert_yaxis()
plt.xlabel("Expected Assists (xAG)")
plt.title("Top 3 Expected Assists (xAG)", fontsize=14, fontweight='bold')


for i, v in enumerate(top_xAG['xAG']):
    plt.text(v + 0.1, i, str(v), va='center', fontsize=10, fontweight='bold')

plt.show()
```

```python
import matplotlib.pyplot as plt


top_PPA = df_mf[['Player', 'Squad', 'PPA']].sort_values(by='PPA', ascending=False).head(3)


plt.figure(figsize=(8,5))
plt.barh(top_PPA['Player'] + " (" + top_PPA['Squad'] + ")", top_PPA['PPA'], color='seagreen')

plt.gca().invert_yaxis()
plt.xlabel("Passes into Penalty Area (PPA)")
plt.title("Top 3 Passers into the Penalty Area (PPA)", fontsize=14, fontweight='bold')

for i, v in enumerate(top_PPA['PPA']):
    plt.text(v + 1, i, str(v), va='center', fontsize=10, fontweight='bold')

plt.show()
```
